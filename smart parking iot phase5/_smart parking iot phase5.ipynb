```python
class ParkingLot:
    def __init__(self, capacity):
        self.capacity = capacity
        self.available_spaces = capacity

    def park_vehicle(self):
        if self.available_spaces > 0:
            self.available_spaces -= 1
            return True
        else:
            return False

    def leave_parking(self):
        if self.available_spaces < self.capacity:
            self.available_spaces += 1
            return True
        else:
            return False

def check_parking_availability(parking_lot):
    return parking_lot.available_spaces

def display_parking_lot(parking_lot):
    car = "🚗"
    empty_space = "🅿️"

    parking_spaces = [car] * (parking_lot.capacity - parking_lot.available_spaces) + [empty_space] * parking_lot.available_spaces

    for i in range(0, len(parking_spaces), 10):
        print(" ".join(parking_spaces[i:i+10]))

def smart_parking_app():
    parking_lot = ParkingLot(20)  # Initialize a parking lot with 20 spaces

    while True:
        print("\nSmart Parking Simulation")
        display_parking_lot(parking_lot)
        print("1. Park Vehicle")
        print("2. Remove Vehicle")
        print("3. Check Availability")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            if parking_lot.park_vehicle():
                print("Vehicle parked.")
            else:
                print("Parking full.")
        elif choice == "2":
            if parking_lot.leave_parking():
                print("Vehicle removed.")
            else:
                print("No vehicles to remove.")
        elif choice == "3":
            availability = check_parking_availability(parking_lot)
            print(f"Available parking spaces: {availability}/{parking_lot.capacity}")
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please enter a valid option.")

smart_parking_app()
```
